# Fremont bridge time series (Visualization)

The Fremont Bridge Bicycle Counter began operation in October 2012 and records the number of bikes that cross the bridge using the pedestrian/bicycle pathways. Inductive loops on the east and west pathways count the passing of bicycles regardless of travel direction. The data consists of a date/time field: Date, east pathway count field: Fremont Bridge NB, and west pathway count field: Fremont Bridge SB. The count fields represent the total bicycles detected during the specified one hour period. Direction of travel is not specified, but in general most traffic in the Fremont Bridge NB field is travelling northbound and most traffic in the Fremont Bridge SB field is travelling southbound.

```python
import numpy as np
import pandas as pd
```

## loading the data

```python
data = pd.read_csv("./data/fremont-bridge/fremont-bridge.csv", index_col= 'Date', parse_dates=True)
data
```

```python
# for convenince we will add east and west to a new columns
data.rename(columns={'Fremont Bridge Total':'Total',
                    'Fremont Bridge East Sidewalk': 'East',
                    'Fremont Bridge West Sidewalk': 'West'}, inplace=True)
data.head()
```

## Describing data

```python
data.dropna().describe()
```

## Visualizing the data
By visualization we can get some insight. Let's plot some raw data.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
data.plot()
plt.ylabel('Hourly Bicycle count')
plt.show()
```

```python
# Managing this data with hours may be cumersome 
# so we can sample data by weeks

weekly = data.resample("W").sum()
weekly.plot(style=[':', '--', '-'])
plt.ylabel('Weekly bicycle count')
plt.show();
```

This shows us some intresting tends that we might expect people bicycle more in summer than winters, and even within a particular season the bicycle use varies from week to week.

```python
# Another way that comes handy for aggregation the data is to use a rolling mean. 
# Here, we will do 30 day rolling mean of our data.

daily = data.resample("D").sum()
daily.rolling(30, center=True).sum().plot(style=[':', '--', '-'])
plt.ylabel('mean hourly count')
plt.show()
```

```python
# The jaggedness is due to the hard cutoff of the window. We can get smoother version
# of rolling mean using a windown function - for example gaussian window

daily.rolling(50, win_type='gaussian', center=True).sum(std=10).plot(style=[':', '--', '-'])
plt.show()
```

```python
# While smoothed data views are useful to get an idea of the genral trend, they hide much of the
# intresting structure. For example we might want to look the average traffice as funtion of the
# time of the day.

by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=hourly_ticks, style=[':', '--', '-'])
plt.ylabel('Traffic accourding to time')
plt.show()
```
